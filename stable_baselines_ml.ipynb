```python
import gym
import gym_fishing
import numpy as np

from stable_baselines.deepq.policies import MlpPolicy
from stable_baselines import TD3, SAC, DDPG, DQN, ACER
from stable_baselines.ddpg.noise import NormalActionNoise, OrnsteinUhlenbeckActionNoise
from stable_baselines.common import make_vec_env

env = gym.make('fishing-v0', n_actions=100, sigma=0.1)
```

```python
model = DQN(MlpPolicy, env, verbose=1, prioritized_replay=True, learning_starts=0)
```

```python
model.learn(total_timesteps=100000)
```

```python
import bokeh.io
from itertools import count
import bokeh.plotting
from bokeh.models import Span

bokeh.io.output_notebook()
```

```python
n_samples = 10
# Set up plots
plot = bokeh.plotting.figure(plot_width=600,
                               plot_height=400,
                               x_axis_label='time',
                               y_axis_label='N')


env = gym.make('fishing-v0', n_actions=100, sigma=0.1)
rewards = []
# 
for sample in range(n_samples):
    # import pdb;pdb.set_trace()
    obs = env.reset()
    sample_traj = [env.fish_population[0]]
    count = 0
    for season in range(1000):
        action, _ = model.predict(obs)
        obs, reward, _, _ = env.step(action)
        sample_traj.append(env.fish_population[0])
        count += reward
    sample_traj = np.array(sample_traj)
    rewards.append(count)
    plot.line(range(len(sample_traj)), sample_traj, line_width=0.5, alpha=0.5)


bokeh.io.show(plot)
```
